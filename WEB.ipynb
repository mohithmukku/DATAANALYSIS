```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.collegedekho.com/news/gate-2023-toppers-list-2023-live-updates-result-link-air-marks-37771/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
```

```python
for i in rows[1:]:
  data =i.find_all("td")
  row =[tr.text for tr in data]
  l = len(df)
  df.loc[l] =row
```

```python
df
```

```python
df.to_csv("toppers.csv")
```
